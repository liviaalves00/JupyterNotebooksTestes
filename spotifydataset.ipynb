```python
genre_series = pd.Series(spo_dt['genre'])
genre_counts = genre_series.value_counts()

def maior_genero_qtd(genre_counts):
    if genre_counts.nunique() == 1:
        return 'valores iguais'
    else:
        max_genre = genre_counts.idxmax()
        max_count = genre_counts.max()
        return max_genre, max_count

resultado = maior_genero_qtd(genre_counts)
print(resultado)
```
